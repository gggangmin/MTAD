```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('./data.pkl','rb') as f:
    data = pickle.load(f)
```

```python
data[:20]
```

```python
# ng인 데이터 10 sequence 정리
ng_index = np.where(data['TAG']=='NG')
```

```python
ng_index
```

```python
temp_ng = []
speed_ng = []
weight_ng = []
index_ng = []

for ngi in ng_index[0]:
    temp_ng.append(list(data['MELT_TEMP'][ngi-10:ngi]/10))
    speed_ng.append(list(data['MOTORSPEED'][ngi-10:ngi]/10))
    weight_ng.append(list(data['MELT_WEIGHT'][ngi-10:ngi]))
    index_ng.append(ngi)
```

```python
for i in range(len(temp_ng)-1,0,-1):
    fig,ax = plt.subplots()
    ax.plot(temp_ng[i],label='temp')
    ax.plot(speed_ng[i],label='speed')
    ax.plot(weight_ng[i],label='weight')
    print(index_ng[i])
    plt.ylim(0,600)
    plt.legend()
    plt.show()
    if i==len(temp_ng)-50:
        break
```
